```python
scores = {"Alice": 95, "Bob": 87, "Charlie": 92}
total=0
for key in scores:
    print(f"{key}: {scores[key]}")
    print("***")
    print(key)
    if key=="Bob":
        break
    total += scores[key]
    print(total)
```

```python
scores = {"Alice": 95, "Bob": 87, "Charlie": 92}
total=0
for values in scores.values():
    print(f"values : {values}   ")
    if values < 90: 
        break
    print("***")
    total += values
    print(f"total : {total}   ")
```
