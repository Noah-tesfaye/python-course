```python
class Car:
    def __init__(self,price):
        self.price=30000
```

```python
class Backpack:
    def __init__(self,price,size):
        self.items=[]
        self.price=price
        self.size=size
        self.type="cloth"
```

```python
class Circle:
    def __init__(self,radius):
        self.radius=radius
```

```python
class Rectangle:
    def __init__(self,length,width):
        self.length=length
        self.width=width
```

```python
class Movie:
    def __init__(self,topic,rating,title,producer):
        self.topic=topic
        self.title=title
        self.rating=rating
        self.producer=producer
```
