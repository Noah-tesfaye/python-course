```python
age=input("How old are you? ")
if int(age) > 100:
    print("You are too old to drive.")
elif int(age) >= 16:
    print("You are ready to get your driver's license  .")
elif int(age) == 15:
    print("You can start learning to drive with a permit.")
elif int(age) == 14:
    print("You can get a learner's permit.")
elif 0 < int(age) < 14:
    print("You are not old enough to drive yet.")
else:
    print("Please enter a valid age.")
```
